```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(t):
    return np.sin(2*np.pi*t)
t1 = np.arange(0,5,0.1)
t2 = np.arange(0,5,0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1),'yo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
x = np.linspace(0,np.pi*6,100)
fig = plt.figure() # generate Figure
ax = fig.add_subplot(111) # generate Axes
line1, = ax.plot(x,np.sin(x),label="1st plot",alpha = 0.2) # add Line2D with a couple of detail information  
line2, = ax.plot(x,np.cos(x),label= "2nd plot")
ax.legend()
print('line1:\n',line1)
```

```python
import random
x = np.linspace(0,2*np.pi,100)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,np.sin(x),"r--",label="sin(x)")
ax1 = ax.twinx()
ax1.plot(x,2*np.sin(x+np.pi/4),label="sin(x+4pi")
ax.scatter(x,np.random.rand(len(x)),label = "scatter")
ax.legend()
ax1.legend()
ax.set_xlabel('x value')
ax.set_ylabel('y value')
plt.show()
```
